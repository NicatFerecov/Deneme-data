```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/StudentsPerformance (1).csv')
```

```python
df.info()
```

```python
df.head(6)
```

```python
new_data = df[df['math score'] == 100]
```

```python
new_data
```

```python
plt.figure(figsize=(6,6))
new_data['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution of Those with a Math Score of 100")
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(new_data['reading score'], new_data['writing score'], color='blue')
plt.title("Math 100 Reading and Writing Scores by Field")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

sns.set(style="whitegrid")  # modern ve temiz tema
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(new_data['reading score'], shade=True, label='Reading Score')
sns.kdeplot(new_data['writing score'], shade=True, label='Writing Score')
plt.title("Math 100 Reading and Writing Score Distribution by Field")
plt.xlabel("Score")
plt.ylabel("Yoğunluk")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df, x='math score', hue=df['math score'] == 100, multiple='stack', palette='Set2')
plt.title("Distribution of Students Who Scored 100 in Math and Those Who Did Not")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(new_data[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Math 100 Field Score Correlation Matrix")
plt.show()
```
